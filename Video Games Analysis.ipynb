# EXPLORATORY DATA ANALYSIS :

## DATA VISULIZATIONS

`Now we will see which Genre has the most counts`

```python
sns.set(rc={'figure.figsize':(25,6)})
ax = sns.countplot(data = data, x = 'GENRE')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count VS Genre",size=20,color='b')

plt.show()
```

`From above graph we can see that 'Action' has the highest counts and 'Board Game' has lowest`

`Which console were most used ?`

```python
sns.set(rc={'figure.figsize':(25,6)})
ax = sns.countplot(data = data, x = 'CONSOLE',palette="deep")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count VS Console",size=20,color='m')
plt.show()
```

`As per the above results , 'X360' were used most`

`We will analyze which game had made the highest sales`

```python
TOTAL_SALES=data[['TOTAL_SALES', 'TITLE']].groupby(['TITLE']).median().sort_values("TOTAL_SALES", ascending = False)
ax = sns.catplot(x="TOTAL_SALES", y=TOTAL_SALES.head(10).index.tolist(), data=TOTAL_SALES.head(10),kind="bar",palette="flare",height=8)
#plt.gca().set_yticklabels([])
plt.xticks(rotation=90)
#plt.yticks(visible=False)
plt.grid(False)
plt.show()
```

`GTAV had been the most purchased game!!!`

```python
TOTAL_SALES=data[['TOTAL_SALES', 'GENRE']].groupby(['GENRE']).median().sort_values("TOTAL_SALES", ascending = False)
ay = sns.catplot(x="TOTAL_SALES", y=TOTAL_SALES.head(10).index.tolist(), data=TOTAL_SALES.head(10),kind="bar",palette="flare",height=8)
#plt.gca().set_yticklabels([])
plt.xticks(rotation=90)
#plt.yticks(visible=False)
plt.grid(False)
plt.show()
```
